```python
import matplotlib.pyplot as plt

# Creating data for plotting
intermediate_data = movies.drop(columns=['movie_id', 'video_release_date', 'title', 'release_date', 'IMDb_URL'], axis=1)
statistic = {}

for i in intermediate_data:
    statistic[i] = intermediate_data[i].sum()

genres = list(statistic.keys())
values = list(statistic.values())

# Creating the bar plot
fig = plt.figure(figsize = (20, 20))
 
plt.bar(genres, values, color ='maroon', 
        width = 0.4)
 
plt.xlabel("Genres of movies")
plt.ylabel("No. of movies")
plt.title("Quantity of genres in data")
plt.show()
```

# Data preparation

## Merging data to create dataset for learning

```python
movies = movies.drop(columns=['IMDb_URL', 'video_release_date', 'title', 'release_date'])

user_preferences = pd.merge(ratings, users, on='user_id')
user_preferences = pd.merge(user_preferences, movies, on='movie_id')
user_preferences = user_preferences.drop(columns=['timestamp', 'zip_code'], axis=1)
```

```python
user_preferences
```

## Labeling string data

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Fit and transform the 'occupation' column
user_preferences['occupation_encoded'] = le.fit_transform(user_preferences['occupation'])
user_preferences['gender_encoded'] = le.fit_transform(user_preferences['gender'])

user_preferences = user_preferences.drop(columns=['occupation'], axis=1)
user_preferences = user_preferences.drop(columns=['gender'], axis=1)
```
